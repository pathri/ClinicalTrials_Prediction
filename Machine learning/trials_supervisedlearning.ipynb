```python
from sklearn.utils import shuffle
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

data =df1
data =data.dropna().reset_index(drop = True)

#x,y =shuffle(data.iloc[:, 2:17],data.iloc[:,1], random_state=98)
x,y =shuffle(data.iloc[:, 2:],data.iloc[:,1], random_state=98)

scaled_x= scaler.fit_transform(x)
x_train, x_test, y_train, y_test =  train_test_split(scaled_x,y,test_size=0.2, random_state=42)
```

```python
#with parameter tuning

def GBDT_tuning(x,y):   
    parameters = {'n_estimators':[100, 200,300,400,500],'learning_rate':[0.1,0.05,0.2],'max_depth':[10,15,20]}


    print("# Tuning hyper-parameters ")
    print()
    clf = GridSearchCV(GradientBoostingClassifier(), parameters, cv=5,
                           scoring= 'roc_auc')
    clf.fit(x, y)
    print("Best parameters set found on development set:", clf.best_params_)
    print("Best cv_results_ set found on development set:", clf.cv_results_)
    return clf.best_params_,clf.cv_results_
GBDT_tuning(x_train,y_train)

##Best parameters set found on development set n=5: {'learning_rate': 0.2, 'max_depth': 10, 'n_estimators': 100}

##Best parameters set found on development set n=100: {'learning_rate': 0.1, 'max_depth': 10, 'n_estimators': 200}

##Best parameters set found on development set n= 50,10: {'learning_rate': 0.1, 'max_depth': 10, 'n_estimators': 200}
```

```python
#with parameter tuning for RF

def RF_tuning(x,y):

    parameters = {'n_estimators':[200,300,400,500],'max_depth':[10,15,20]}
    print("# Tuning hyper-parameters rf")
    print()
    clf = GridSearchCV(RandomForestClassifier(), parameters, cv=5,
                           scoring= 'roc_auc')
    clf.fit(x, y)
    print("Best parameters set found on development set rf",clf.best_params_)
    return clf.best_params_,clf.cv_results_

RF_tuning(x_train,y_train)
#Best parameters set found on development set n=5: {'max_depth': 15, 'n_estimators': 500}

##Best parameters set found on development set n= 100,50,10 rf {'max_depth': 10, 'n_estimators': 200}

###Best parameters set found on development set rf {'max_depth': 10, 'n_estimators': 200}
```

```python
#with parameter tuning for svm

def SVM_tuning(x,y):

    parameters = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf']}
    print("# Tuning hyper-parameters svm")
    print()
    clf = GridSearchCV(SVC(), parameters, cv=5,
                           scoring= 'roc_auc')
    clf.fit(x, y)
    print("Best parameters set found on development set svm :",clf.best_params_)
    return clf.best_params_,clf.cv_results_

SVM_tuning(x_train,y_train)
#Best parameters set found on development set n=5: {'C': 100, 'gamma': 1, 'kernel': 'rbf'} n=5
#Best parameters set found on development set svm : {'C': 0.1, 'gamma': 1, 'kernel': 'rbf'} n=100
#Best parameters set found on development set svm : {'C': 0.1, 'gamma': 1, 'kernel': 'rbf'} n=50,10
```

```python
from sklearn.metrics import confusion_matrix

clf_GBT =GradientBoostingClassifier(n_estimators=100,max_depth=10, learning_rate=0.2)#n=3


#clf_GBT =GradientBoostingClassifier(n_estimators=10,max_depth=10, learning_rate=0.1)#n=5
#lf_GBT =GradientBoostingClassifier(n_estimators=200,max_depth=10, learning_rate=0.1) #n=100,50
clf_GBT = clf_GBT.fit(x_train,y_train)
pre_y_list = clf_GBT.predict(x_test)  
pre_y_proba = clf_GBT.predict_proba(x_test)
fpr_GBT,tpr_GBT, _GBT = roc_curve(y_test, pre_y_proba[:,1],pos_label=1)
auc_GBT = auc(fpr_GBT, tpr_GBT)
accuracy_score_GBT=accuracy_score(y_test,pre_y_list)
precision_score_GBT = precision_score(y_test,pre_y_list)
recall_score_GBT = recall_score(y_test,pre_y_list)
f1_score_GBT  = f1_score(y_test,pre_y_list)

print(auc_GBT,accuracy_score_GBT,precision_score_GBT,recall_score_GBT,f1_score_GBT )
print(confusion_matrix(y_test, pre_y_list))
```

```python
##Best parameters set found on development set: {'max_depth': 15, 'n_estimators': 500}

#clf_RF =RandomForestClassifier(n_estimators=10,max_depth=20,random_state = 25) #dryrun
#clf_RF =RandomForestClassifier(n_estimators=500,max_depth=15) #n=5
clf_RF =RandomForestClassifier(n_estimators=200,max_depth=10) #n=5

#clf_RF =RandomForestClassifier(n_estimators=200,max_depth=10) #n=100,50
clf_RF = clf_RF.fit(x_train,y_train)
pre_y_RF = clf_RF.predict(x_test)  
pre_y_proba_RF = clf_RF.predict_proba(x_test)
fpr_RF, tpr_RF, _RF = roc_curve(y_test, pre_y_proba_RF[:,1],pos_label=1)
auc_RF = auc(fpr_RF, tpr_RF)
accuracy_score_RF=accuracy_score(y_test,pre_y_RF)
precision_score_RF = precision_score(y_test,pre_y_RF)
recall_score_RF = recall_score(y_test,pre_y_RF)
f1_score_RF  = f1_score(y_test,pre_y_RF)
print (auc_RF,accuracy_score_RF,precision_score_RF,recall_score_RF,f1_score_RF )
print(confusion_matrix(y_test, pre_y_RF))
```
